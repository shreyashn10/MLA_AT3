### D.3 New Feature "tenure_band"

> This feature transforms raw tenure into easy-to-read loyalty stages, which helps group customers by lifecycle phase in a meaningful way.

```python
def assign_tenure_band(months):
    if months < 12:
        return 'New'
    elif 12 <= months < 36:
        return 'Mid'
    elif 36 <= months <= 60:
        return 'Loyal'
    else:
        return 'Long-Term'
```

```python
df_eng['tenure_band'] = df_eng['month_tenure'].apply(assign_tenure_band)
```

```python
df_eng['tenure_band'].value_counts()
```

```python
# <Student to fill this section>
feature_engineering_2_explanations = """
This feature groups customers by how long they’ve been with the company. It helps the Product Team distinguish newer users from those with deeper product experience. Each group may require a different digital upsell approach based on familiarity.
"""
```
