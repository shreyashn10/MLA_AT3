### C.3 Fixing "outliers in total_revenue"

> Provide some explanations on why you believe it is important to fix this issue and its impacts

> You can add more cells in this section

```python
# <Student to fill this section>
column_to_cap = 'total_revenue'

print(f"--- Handling Outliers for: {column_to_cap} ---")

# Ensure the column exists and is numeric
if column_to_cap not in df_clean.columns:
    print(f"Column '{column_to_cap}' not found in df_clean. Skipping outlier handling for this column.")
elif not pd.api.types.is_numeric_dtype(df_clean[column_to_cap]):
    print(f"Column '{column_to_cap}' is not numeric. Skipping outlier handling.")
else:
    print(f"\nDescriptive statistics for {column_to_cap} BEFORE capping:")
    print(df_clean[column_to_cap].describe())

    # 1. Calculate IQR and bounds
    Q1 = df_clean[column_to_cap].quantile(0.25)
    Q3 = df_clean[column_to_cap].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"\nCalculated for {column_to_cap}:")
    print(f"  Q1: {Q1:.2f}")
    print(f"  Q3: {Q3:.2f}")
    print(f"  IQR: {IQR:.2f}")
    print(f"  Lower Bound (for capping): {lower_bound:.2f}")
    print(f"  Upper Bound (for capping): {upper_bound:.2f}")

    # 2. Identify outliers before capping
    outliers_low = df_clean[df_clean[column_to_cap] < lower_bound]
    outliers_high = df_clean[df_clean[column_to_cap] > upper_bound]

    print(f"\nNumber of potential outliers in {column_to_cap} BEFORE capping:")
    print(f"  Below lower bound ({lower_bound:.2f}): {len(outliers_low)}")
    print(f"  Above upper bound ({upper_bound:.2f}): {len(outliers_high)}")

    # 3. Apply capping to df_clean
    # Using .loc to ensure modification of the original df_clean and avoid SettingWithCopyWarning
    df_clean.loc[df_clean[column_to_cap] < lower_bound, column_to_cap] = lower_bound
    df_clean.loc[df_clean[column_to_cap] > upper_bound, column_to_cap] = upper_bound

    print(f"\nCapping applied to {column_to_cap} in df_clean.")

    # 4. Show statistics AFTER capping
    print(f"\nDescriptive statistics for {column_to_cap} AFTER capping:")
    print(df_clean[column_to_cap].describe())

    # Verify if outliers were capped
    outliers_low_after = df_clean[df_clean[column_to_cap] < lower_bound]
    outliers_high_after = df_clean[df_clean[column_to_cap] > upper_bound]

    print(f"\nNumber of values outside original bounds in {column_to_cap} AFTER capping:")
    print(f"  Below original lower bound ({lower_bound:.2f}): {len(outliers_low_after)}")
    print(f"  Above original upper bound ({upper_bound:.2f}): {len(outliers_high_after)}")

    # Count how many values are now exactly at the bounds
    at_lower_bound = df_clean[df_clean[column_to_cap] == lower_bound].shape[0]
    at_upper_bound = df_clean[df_clean[column_to_cap] == upper_bound].shape[0]
    print(f"  Number of values at new lower bound: {at_lower_bound}")
    print(f"  Number of values at new upper bound: {at_upper_bound}")
```
